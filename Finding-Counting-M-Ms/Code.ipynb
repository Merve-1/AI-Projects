```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from pathlib import Path
```

```python
img = cv2.imread('Images/0.png')
imageName = Path('Images/0.png').stem
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
#yellow
yLower = np.array([19, 90, 70],np.uint8)
yUpper = np.array([30, 255, 255], np.uint8)
yMask = cv2.inRange(img, yLower, yUpper)

#red
rLower = np.array([160,170,100],np.uint8)
rUpper = np.array([180,255,255],np.uint8)
rMask = cv2.inRange(img, rLower, rUpper)

#green 
gLower = np.array([30,50,100],np.uint8)
gUpper = np.array([70,255,155],np.uint8)
gMask = cv2.inRange(img, gLower, gUpper)

#blue
bLower = np.array([88,80,100],np.uint8)
bUpper = np.array([120,255,255],np.uint8)
bMask = cv2.inRange(img, bLower, bUpper)

#orange
oLower = np.array([1,110,195],np.uint8)
oUpper = np.array([20,230,255],np.uint8)
oMask = cv2.inRange(img, oLower, oUpper)

#brown 
brLower = np.array([10, 100, 5],np.uint8)
brUpper = np.array([180, 255, 30],np.uint8)
brMask = cv2.inRange(img, brLower, brUpper)


```

```python
img = cv2.cvtColor(img,cv2.COLOR_HSV2RGB)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
out = cv2.GaussianBlur(img,ksize=(25,25),sigmaX= 5)
plt.imshow(out)
```

```python
canny_img= cv2.Canny(image=out,threshold1=71,threshold2=86)
plt.imshow(canny_img)
```

```python
#Hough Transformation
circles = cv2.HoughCircles(canny_img, cv2.HOUGH_GRADIENT, 1,  100, param1=200,
                           param2=10, minRadius=30, maxRadius=65)
```

```python
counter = 0
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # Draw outer circle
        cv2.circle(img, (i[0], i[1]), 53, (255, 0, 255), 7)
        counter = counter + 1
```
